```python

from simpleimage import SimpleImage
colour = [[255, 25, 25], [25, 255, 25], [25, 25, 255], [20, 150, 250]]
def main():
    vertical_stripes = ver_line(50, 10, colour)
    vertical_stripes.show()
    horizontal_stripes = hori_line(50, 20, colour)
    horizontal_stripes.show()
    diagonal_stripes1 = diag_line1(50, 10,colour)
    diagonal_stripes1.show()
    diagonal_stripes2 = diag_line2(50, 10,colour)
    diagonal_stripes2.show()
    vertical_daimonds1 = ver_dai(50, 10, colour)
    vertical_daimonds1.show()
    vertical_daimonds2 = ver_dai2(50, 10, colour)
    vertical_daimonds2.show()
    zigzag_daimond1 = zig_dai1(50, 10, colour)
    zigzag_daimond1.show()
    zigzag_daimond2 = zig_dai2(50, 10, colour)
    zigzag_daimond2.show()

# "Image with vertical stripes "
def ver_line(col_width, rep, colour):
    image = SimpleImage.blank(col_width * rep * len(colour), col_width * rep * len(colour))
    p = 0
    n = 0
    for y in range(col_width * rep * len(colour)):
        for x in range(col_width * rep * len(colour)):
            pixel = image.get_pixel(x,y)
            #To change colour of column
            if n == col_width:
                n -= col_width
                p += 1
                if p == len(colour):
                    p -= len(colour)
            pixel.red = colour[p][0]
            pixel.green = colour[p][1]
            pixel.blue = colour[p][2]
            n += 1
    return image

# "Image with horizontal stripes"
def hori_line(row_height, no_rows, colour):
    image = SimpleImage.blank(row_height * no_rows, row_height * no_rows)
    p = 0
    n = 0
    r = 0
    for y in range(row_height * no_rows):
        #To change colour of row
        n += 1
        if n == row_height:
            n -= row_height
            p += 1
            if p == len(colour):
                p -= len(colour)
        for x in range(row_height * no_rows):
            pixel = image.get_pixel(x,y)
            pixel.red = colour[p][0]
            pixel.green = colour[p][1]
            pixel.blue = colour[p][2]
    return image

#Image with diagonal stripes
def diag_line1(col_width, rep, colour):
    image = SimpleImage.blank(col_width * rep * len(colour) - 1, col_width * rep * len(colour) - 1)
    p = 0
    n = 0
    for y in range(col_width * rep * len(colour) - 1):
        for x in range(col_width * rep * len(colour) - 1):
            pixel = image.get_pixel(x,y)
            #To change colour of column
            if n == col_width:
                n -= col_width
                p += 1
                if p == len(colour):
                    p -= len(colour)
            pixel.red = colour[p][0]
            pixel.green = colour[p][1]
            pixel.blue = colour[p][2]
            n += 1
    return image


def diag_line2(col_width, rep, colour):
    image = SimpleImage.blank(col_width * rep * len(colour) - 1, col_width * rep * len(colour) - 1)
    p = 0
    n = 0
    r = 0
    for y in range(col_width * rep * len(colour) - 1):
        #To start fresh on every 100th row
        r += 1
        if r == col_width + 1:
            r -= col_width
            n = 0
            p += 1
            if p >= len(colour):
                p = p - len(colour)
        for x in range(col_width * rep * len(colour) - 1):
            pixel = image.get_pixel(x,y)
            #To change colour of column
            if n == col_width:
                n -= col_width
                p += 1
                if p == len(colour):
                    p -= len(colour)
            pixel.red = colour[p][0]
            pixel.green = colour[p][1]
            pixel.blue = colour[p][2]
            n += 1
    return image


#Image with vertical colour daimonds
def ver_dai1(col_width, rep, colour):
    image = SimpleImage.blank(col_width * rep * len(colour) - 1, col_width * rep * len(colour) - 1)
    p = 0
    n = 0
    r = 0
    for y in range(col_width * rep * len(colour) - 1):
        #To start fresh on every 100th row
        r += 1
        if r == col_width + 1:
            r -= col_width
            n = 0
            p = 0
        for x in range(col_width * rep * len(colour) - 1):
            pixel = image.get_pixel(x,y)
            #To change colour of column
            if n == col_width:
                n -= col_width
                p += 1
                if p == len(colour):
                    p -= len(colour)
            pixel.red = colour[p][0]
            pixel.green = colour[p][1]
            pixel.blue = colour[p][2]
            n += 1
    return image


def ver_dai2(col_width, rep, colour):
    image = SimpleImage.blank(col_width * rep * len(colour) - 1, col_width * rep * len(colour) - 1)
    p = 0
    n = 0
    r = 0
    for y in range(col_width * rep * len(colour) - 1):
        #To start fresh on every 100th row
        r += 1
        if r == col_width + 1:
            r -= col_width
            n = 0
            p += 2
            if p >= len(colour):
                p = p - len(colour)
        for x in range(col_width * rep * len(colour) - 1):
            pixel = image.get_pixel(x,y)
            #To change colour of column
            if n == col_width:
                n -= col_width
                p += 1
                if p == len(colour):
                    p -= len(colour)
            pixel.red = colour[p][0]
            pixel.green = colour[p][1]
            pixel.blue = colour[p][2]
            n += 1
    return image

#Image with diagonal colour daimonds
def zig_dai1(col_width, rep, colour):
    image = SimpleImage.blank(col_width * rep * len(colour) - 1, col_width * rep * len(colour) - 1)
    p = 0
    n = 0
    r = 0
    c = 0
    for y in range(col_width * rep * len(colour) - 1):
        #To start fresh on every 100th row
        r += 1
        if r == col_width + 1:
            r -= col_width
            n = 0
            c += 1
            p = c % len(colour)
        for x in range(col_width * rep * len(colour) - 1):
            pixel = image.get_pixel(x,y)
            #To change colour of column
            if n == col_width:
                n -= col_width
                p += 1
                if p == len(colour):
                    p -= len(colour)
            pixel.red = colour[p][0]
            pixel.green = colour[p][1]
            pixel.blue = colour[p][2]
            n += 1
    return image

def zig_dai2(col_width, rep, colour):
    image = SimpleImage.blank(col_width * rep * len(colour) - 1, col_width * rep * len(colour) - 1)
    p = 0
    n = 0
    r = 0
    for y in range(col_width * rep * len(colour) - 1):
        #To start fresh on every 100th row
        r += 1
        if r == col_width + 1:
            r -= col_width
            n = 0
            p += 3
            if p >= len(colour):
                p = p - len(colour)
        for x in range(col_width * rep * len(colour) - 1):
            pixel = image.get_pixel(x,y)
            #To change colour of column
            if n == col_width:
                n -= col_width
                p += 1
                if p == len(colour):
                    p -= len(colour)
            pixel.red = colour[p][0]
            pixel.green = colour[p][1]
            pixel.blue = colour[p][2]
            n += 1
    return image


if __name__ == '__main__':
    main()
```
